```python
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.contour3D(X,Y,Z,50,cmap='viridis')
ax.set_title('Simple 3D contour plot')
plt.show()
```

```python
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.plot_surface(X,Y,Z,cmap='viridis')
ax.set_title('Simple 3D surface plot')
plt.show()
```

```python
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.plot_wireframe(X,Y,Z,color='blue')
ax.set_title('Simple 3D wireframe plot')
plt.show()
```
